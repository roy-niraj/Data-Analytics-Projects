```python
# Remove brackets and quotes properly
df["product_category_tree"] = df["product_category_tree"].str.replace(r'[\[\]"]', '', regex=True)

# Extract main category correctly
df["main_category"] = df["product_category_tree"].str.split(">>").str[0].str.strip()
```

```python
df = df[df["main_category"].str.len() < 50]
```

```python
category_counts = df["main_category"].value_counts()
valid_categories = category_counts[category_counts > 10].index

df = df[df["main_category"].isin(valid_categories)]
```

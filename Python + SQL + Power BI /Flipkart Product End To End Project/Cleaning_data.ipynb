```python
df = df.dropna(how = 'all')
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns = ['product_url', 'image', 'description', 'product_specifications', 'crawl_timestamp'])
```

```python
df.head(5)
```

```python
df['brand'] = df['brand'].fillna('unknow')
```

```python
df['retail_price'] = pd.to_numeric(df['retail_price'], errors = 'coerce')
```

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors = 'coerce')
```

```python
df['product_rating'] = pd.to_numeric(df['product_rating'], errors = 'coerce')
df['overall_rating'] = pd.to_numeric(df['product_rating'], errors = 'coerce')
```

```python
df["overall_rating"] = df["overall_rating"].fillna(0)
df["product_rating"] = df["product_rating"].fillna(0)
```

```python
df['main_category'] = df['product_category_tree'].str.split('>>').str[0]
```

```python
df.shape
```

```python
df = df.dropna(subset=["retail_price", "discounted_price"])
```
